## Advection-Diffusion Equation <a id="advection_diffusion_equation" />

Consider the classic advection-diffusion equation given by

$$ u_t + v u_x = \alpha u_{xx}, $$

where $v$ could be a superficial velocity, $\alpha$ a diffusion coefficient, and $u$ the concentration of some species. 

### Pure Advection

For $\alpha=0$, we know the solution:

$$ u(t,z) = u_0(z-vt). $$

Let's use it to explore some numerical phenomena!

```julia
# Imports
using LinearAlgebra, Plots, LaTeXStrings

# Defining the advection diffusion equation
function adv_diff_ODE(u, p, t)

    uBCs, v, α, A_1D_CDS, A_1D_UDS = p

    uzg_vec = copy(u)
    insert!(uzg_vec, 1, uBCs[1])
    push!(uzg_vec, u[end])

    du = ( -v.*A_1D_UDS*uzg_vec + α.*A_1D_CDS*uzg_vec )[2:end-1]

end

# Defining a simple ODE solver
function solve_ODE(ODE_fun, t_span, u0, t_eval, dt, p)

    t0, tf = t_span
    u = u0

    utz_mat = zeros(length(t_eval), length(u))
    utz_mat[1,:] = u
    i_teval = 2
    t = t0
    
    while t < tf
        du = ODE_fun(u, p, t)
        u += du * dt
        t += dt
        if t > t_eval[i_teval]
            utz_mat[i_teval,:] = copy(u)
            i_teval += 1
        end
    end

    return utz_mat
end
```

```julia
# Problem specs
L = 1.
v = 1.
α = 0.0
uBCs = [1., 0.];        # BCs

# Spacial discretization
Nz = 100
dz = L / (Nz + 1)
uz_vec = ones(Nz);      # IC
A_1D_CDS = 1/(dz^2) .* Tridiagonal(ones(Nz+1), -ones(Nz+2), ones(Nz+1))
A_1D_UDS = 1/dz .* Tridiagonal(-ones(Nz+1), ones(Nz+2), zeros(Nz+1))
A_1D_UDS[end,:] = zeros(Nz+2);

p = uBCs, v, α, A_1D_CDS, A_1D_UDS;
```

```julia
## Test solver for different Courant numbers
Cus = [0.8, 1.0, 1.01]
plts = []

import PyPlot; plt = PyPlot
using PyCall, LaTeXStrings

# line cyclers adapted to colourblind people
cycler = pyimport("cycler").cycler
line_cycler   = (cycler(color=["#E69F00", "#56B4E9", "#009E73", "#0072B2", "#D55E00", "#CC79A7", "#F0E442"]) +
                 cycler(linestyle=["-", "--", "-.", ":", "-", "--", "-."]))
marker_cycler = (cycler(color=["#E69F00", "#56B4E9", "#009E73", "#0072B2", "#D55E00", "#CC79A7", "#F0E442"]) +
                 cycler(linestyle=["none", "none", "none", "none", "none", "none", "none"]) +
                 cycler(marker=["4", "2", "3", "1", "+", "x", "."]))

plt.rc("axes", prop_cycle=line_cycler)
fig, axs = plt.subplots(1,3, figsize=(14, 5))

for (i_Cu,Cu) in enumerate(Cus)
    
    # Set time step to get desired Courant number
    dt = Cu*dz / v

    # Solving
    t_span = [0., 2.]
    u0 = zeros(Nz)
    t_eval = t_span[1]:0.2:t_span[end]
    utz_mat = solve_ODE(adv_diff_ODE, t_span, u0, t_eval, dt, p)

    # Plotting
    axs[i_Cu].plot(dz:dz:L-dz, utz_mat', ls="-", color="black")
    axs[i_Cu].set_xlabel(L"Reactor coordinate, $z$ (m)")
    axs[i_Cu].set_ylabel(L"Velocity, $u$ (m/s)")
end

fig
```

```julia
# Lax-Friedrichs

## Problem specs
L = 1.
v = 1.
α = 0.0
uz_vec = ones(Nz);      # IC
uBCs = [1., 0.];        # BCs

## Spacial discretization
Nz = 100
dz = L / (Nz + 1)

dl = -ones(Nz+1)./2; d = zeros(Nz+2); du = ones(Nz+1)./2
A_1D_LF = 1/dz .* Tridiagonal(dl, d, du);
A_1D_LF
```

```julia
function fun(uz_vec, param_vec)

    uBCs, v, α, dt, A_1D_CDS, A_1D_LF = param_vec

    uzg_vec = copy(uz_vec)
    insert!(uzg_vec, 1, uBCs[1])
    push!(uzg_vec, uz_vec[end])

    uz_vec = ( (uzg_vec[1:end-2] + uzg_vec[3:end])./2 ) + dt * ( -v.*A_1D_LF*uzg_vec )[2:end-1]

    return uz_vec
end


function solve_PDE(fun, t_span, u0, t_eval, param_vec)

    t0, tf = t_span
    uz_vec = u0
    uBCs, v, α, dt, A_1D_CDS, A_1D_LF = param_vec

    utz_mat = zeros(length(t_eval), length(uz_vec))
    utz_mat[1,:] = uz_vec
    i_teval = 2
    t = t0
    
    while t < tf

        uz_vec = fun(uz_vec, param_vec)


        t += dt
        if t > t_eval[i_teval]
            utz_mat[i_teval,:] = copy(uz_vec)
            i_teval += 1
        end
    end

    # Returns solution vector
    return utz_mat
end
```

```julia
Cus = [0.8, 1.0, 1.01]
plots = []
for Cu in Cus
    dt = Cu*dz / v
    param_vec = [uBCs, v, α, dt, A_1D_CDS, A_1D_LF]

    ## Solving
    t_span = [0., 2.]
    u0 = zeros(Nz)
    t_eval = t_span[1]:0.2:t_span[end]
    utz_mat = solve_PDE(fun, t_span, u0, t_eval, param_vec);

    p = plot(dz:dz:L-dz, utz_mat', legend=false, title="Courant: $Cu", xlabel=L"Length, $z$ (m)", ylabel=L"$u(t)$")
    push!(plots, p)
end
```

```julia
# Plotting
plot(plots[1], plots[2], plots[3], layout=(1,3), size=(1100,400))
```

## Burger's Equation

```julia
## Problem specs
L = 20.
u0z_vec = ones(Nz) * 0.3;                    # IC
function uBC(t) 
    if t <= 0
        uBC = 0.3
    elseif t > 0 && t <= 7
        uBC = 0.3 + 0.1*(1-cos((2*pi)/7 * t))
    else
        uBC = 0.3
    end
    return uBC
end

## Spacial discretization
Nz = 200
dz = L / (Nz + 1)

dl = -ones(Nz+1); d = ones(Nz+2); du = zeros(Nz+1)
A_1D_UDS = 1/dz .* Tridiagonal(dl, d, du);
A_1D_UDS[end,:] = zeros(Nz+2);
A_1D_UDS;
```

```julia
function Burgers_1(t, uz_vec, param_vec)

    uBC, dt, A_1D_UDS = param_vec

    uzg_vec = copy(uz_vec)
    insert!(uzg_vec, 1, uBC(t))
    push!(uzg_vec, uz_vec[end])

    uz_vec = uz_vec - uz_vec .* dt .* (A_1D_UDS*uzg_vec)[2:end-1]

    return uz_vec
end


function solve_PDE(fun, t_span, u0, t_eval, param_vec)

    t0, tf = t_span
    uz_vec = u0
    uBCs, dt, A_1D_UDS = param_vec

    t = t0
    utz_mat = zeros(length(t_eval), length(uz_vec)+2)
    utz_mat[1,2:end-1] = copy(uz_vec); utz_mat[1,1] = uBC(t); utz_mat[1,end] = uz_vec[end]
    i_teval = 2
    
    while t < tf

        uz_vec = fun(t, uz_vec, param_vec)

        t += dt
        if t > t_eval[i_teval]
            utz_mat[i_teval,2:end-1] = copy(uz_vec); utz_mat[i_teval,1] = uBC(t); utz_mat[i_teval,end] = uz_vec[end]
            i_teval += 1
        end
    end

    # Returns solution vector
    return utz_mat
end
```

```julia
param_vec = [uBC, dt, A_1D_UDS]

## Solving
t_span = [0., 50.]
t_eval = t_span[1]:5:t_span[end]
utz_mat = solve_PDE(Burgers_1, t_span, u0z_vec, t_eval, param_vec);
```

```julia
plot(0:dz:L, utz_mat', legend=false, title="Burgers method one", xlabel=L"Time, $t$ (s)", ylabel=L"$u(t)$")
```
